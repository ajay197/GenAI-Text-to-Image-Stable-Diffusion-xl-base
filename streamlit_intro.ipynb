```python
%%writefile app.py
import streamlit as st
from PIL import Image

st.title("My Streamlit App")
st.header("This is Header")
st.subheader("This is a subheader")
st.text("This is a text")
st.button("Click Me!")
st.checkbox("Checkbox1")
st.checkbox("Checkbox2")
st.checkbox("Checkbox3")
st.radio("Radio", ['Option1', 'Option2', 'Option3'])
st.file_uploader("File Uploader")
st.text_input("Text Input", placeholder="Enter name")
st.number_input("Number Input", min_value=0, max_value=100, value=50)
st.text_area("Text Area", placeholder="Enter your message here")
st.slider("Slider", min_value=0, max_value=100, value=56)

# Adding Columns
col1, col2 = st.columns(2)
with col1:
    st.header("Column 1")
    st.text("Enter text Col1")
with col2:
    st.header("Column 2")
    st.text("Enter text Col2")

# Add file in uploader and display it
image = st.file_uploader("Upload a file", type=['png', 'jpg', 'jpeg'])
if image:
    st.image(image, caption="Uploaded Image", use_column_width=True)
```
